## Aliasing

A figura abaixo mostra uma onda senoidal de 800 kHz, amostrada a 1 MS/s. A linha pontilhada indica o sinal com **aliasing** registrado nessa taxa de amostragem. O efeito de **aliasing** faz a frequência de 800 kHz ser representada incorretamente como uma onda senoidal de 200 kHz 

image.png

A frequência de **aliasing**, $f_a$, pode ser calculada para sabermos como um sinal de entrada de frequência maior que a frequência de Nyquist será visualizado.

Para calcular a frequência de aliasing devido a amostragem que **não seguiu o critério de Nyquist** é necessário utilizar a equação abaixo:

$$f_a(N)=|(n*f_s)-f_{in}|$$

* **$f_a$:** frequência de alias
* **$f_s$:** taxa de amostragem
* **$f_{in}$:** frequência do sinal de entrada
* **N:** Um inteiro (maior ou igual a 0) que assume um valor necessário para que o termo de $N*f_s$ seja o mais próximo possível da frequência de entrada ($f_{in}$)

Por exemplo, considere uma <ins>frequência de amostragem de 100 Hz</ins> para um sinal de entrada que contém as seguintes frequências: 25Hz, 70Hz, 160Hz e 510Hz. **As frequências <ins>abaixo da frequência de Nyquist (de 50Hz)</ins> serão amostradas corretamente; as frequências acima de 50Hz sofrerão aliasing.**

image.png

Os cálculos das frequências de aliasing são:

$$\text{Aliasing F2} = |1*100-70|=30Hz$$
$$\text{Aliasing F3} = |2*100-160|=40Hz$$
$$\text{Aliasing F4} = |5*100-510|=10Hz$$

Além de aumentar a taxa de amostragem, o aliasing também pode ser evitado pelo uso de um **filtro antialiasing**, um filtro passa-baixas que atenua quaisquer frequências do sinal de entrada maiores que a frequência de Nyquist.

Esse filtro deve ser introduzido antes do conversor AD para evitar que a largura de banda do sinal de entrada viole o critério de amostragem. Os canais de entrada analógica podem ter filtros analógicos digitais implementados no hardware, para ajudar a evitar o aliasing.
